```python
##Step7 - Put everything together - Loop through results and append data inside a list
```

```python
title = []
subtitle = []
author = []
publisher = []
publication_year = []
price = []

for results in results_json:
    title.append(results['title'])
    subtitle.append(results['subtitle'])
    author.append(results['authors'][0]['author_name'])
    publisher.append(results['publisher'])
    publication_year.append(results['publication_year'])
    price.append(results['price'])
```

```python
#Step8 - Pandas Dataframe
```

```python
books_df = pd.DataFrame({'Title':title, 'Subtitle':subtitle, 'Author':author, 'Publiser':publisher, 
                         'Publication_year':publication_year, 'Price':price})
```
